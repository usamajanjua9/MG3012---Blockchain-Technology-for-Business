## Distributed Ledgers & Blockchain Architecture  
### 1️⃣ Distributed Ledger Technology (DLT)  

---

## 📖 Definition  
A **Distributed Ledger** is a database that is **shared, replicated, and synchronized** across multiple participants in a network.  
Unlike traditional databases, there is **no central administrator** — instead, the ledger is maintained collectively.  

---

## 🔑 Core Characteristics  

1. **Decentralized storage of records** 🗂️  
   - Data is not held by one authority → every participant has a copy.  

2. **Replication across nodes** 🌍  
   - Each node maintains its own copy → ensuring redundancy.  

3. **Tamper-resistance (immutability)** 🔒  
   - Once written, data cannot easily be altered or deleted.  

4. **Cryptographic verification** ✅  
   - Transactions are secured with hashing & digital signatures.  

---

## 🆚 DLT vs Centralized Databases  

| Feature              | Centralized DB 🏢                   | Distributed Ledger 🌐   |
|----------------------|--------------------------------------|-------------------------|
| Authority            | One central admin controls all       | Multiple participants   |
| Single Point of Fail | Yes ⚠️                              | No ❌ (redundant copies)|
| Data Integrity       | Trust in central authority           | Verified by cryptography|
| Transparency         | Limited                              | High (shared & auditable)|

👉 Example:  
- **Centralized:** Bank ledger (only the bank updates balances).  
- **DLT:** Bitcoin ledger (any node can verify balances independently).  

---

## 🧾 Types of Distributed Ledgers  

### 🔓 Permissionless (Public)  
- Anyone can join, validate, and participate.  
- Examples: **Bitcoin, Ethereum**.  
- Pros: Open, transparent, censorship-resistant.  
- Cons: Slower, high energy use (PoW), scalability issues.  

### 🔒 Permissioned (Private/Consortium)  
- Restricted access → only selected members can validate transactions.  
- Examples: **Hyperledger Fabric, Corda**.  
- Pros: Faster, efficient, better for enterprises.  
- Cons: Less decentralized, requires governance.  

---

## 💼 Business Benefits of DLT  

- **Transparency:** All participants can verify the ledger.  
- **Trust:** Eliminates need for a single trusted middleman.  
- **Auditability:** Historical records are permanent & verifiable.  
- **Resilience:** Network continues even if some nodes fail.

### 2️⃣ Blockchain Network Model  

---

## 🖥️ Nodes in a Blockchain  

Nodes are **computers that participate in the blockchain network**. Each type of node plays a different role:

1. **Full Nodes** 🏛️  
   - Store the **entire blockchain**.  
   - Validate all transactions & blocks independently.  
   - Enforce consensus rules (reject invalid blocks).  
   - Example: Bitcoin Core node.  

2. **Light Clients (SPV Nodes)** 📱  
   - Store only **block headers** instead of full data.  
   - Use **Simplified Payment Verification (SPV)** to check transactions.  
   - Useful for mobile wallets.  

3. **Miners / Validators** ⛏️  
   - Propose new blocks.  
   - **PoW miners:** solve puzzles (Bitcoin).  
   - **PoS validators:** stake tokens & validate (Ethereum 2.0).  

4. **Archival Nodes vs Pruned Nodes**  
   - **Archival nodes:** Store the **full blockchain history**.  
   - **Pruned nodes:** Store only recent history (saves disk space).

## 🌍 Peer-to-Peer (P2P) Network  

- Blockchain uses a **P2P communication model** → no central server.  
- Nodes connect directly with each other.  
- **Gossip protocol** spreads new transactions & blocks quickly:  
  - Node receives a transaction → shares it with neighbors → spreads across the network like “gossip”.  

📡 **Result:** Everyone eventually receives the same data, without needing a central authority.  

---

## ⏳ Mempool (Memory Pool)  

- **Definition:** Temporary storage for **unconfirmed transactions**.  
- Each node maintains its **own mempool**.  
- Process:  
  1. User creates a transaction.  
  2. Transaction enters mempool (waiting area).  
  3. Miner/validator picks transactions from mempool.  
  4. Once included in a block → transaction leaves mempool.  

📦 **Key Idea:** Mempool is like a **queue of pending payments** before final settlement.

# 🧩 Merkle Root — Step-by-Step

The **Merkle root** is a single hash value that represents **all transactions** inside a block.
It’s built using a **Merkle tree**, which is a special kind of binary tree of hashes.

---

## 🎯 Why Do We Need a Merkle Root?

* A block can have **thousands of transactions**. Storing all of them in the block header would be too large.
* Instead, the block header only stores the **Merkle root** (32 bytes).
* With the Merkle root, anyone can prove a transaction is included in the block using a short proof (log₂(n) hashes).

---

## 🔐 Step 1 — Hashing Transactions

* Each transaction is first turned into a **transaction ID (txid)**.
* A txid is the **double-SHA-256 hash** of the raw transaction data.
* These txids form the **leaves** of the Merkle tree.

👉 Example:

* `tx1 → H1`
* `tx2 → H2`
* `tx3 → H3`
* `tx4 → H4`

---

## 🌳 Step 2 — Pairing and Hashing

* Pair adjacent leaves: `(H1, H2)` and `(H3, H4)`.
* For each pair, concatenate the two hashes (left || right), then hash them together.
* These become the **parent nodes**.

👉 Example:

* `P12 = Hash(H1 || H2)`
* `P34 = Hash(H3 || H4)`

---

## 🌲 Step 3 — Repeat Upwards

* Keep pairing and hashing until only **one hash** remains.
* This final hash is the **Merkle root**.

👉 Example:

* `Merkle Root = Hash(P12 || P34)`

---

## 🔄 Step 4 — Odd Number of Transactions

* If there is an **odd number** of hashes at a level, the **last hash is duplicated**.
* This ensures that every level has an even number of nodes for pairing.

👉 Example:

* If 5 transactions: `H1, H2, H3, H4, H5`
* Pairing: `(H1,H2), (H3,H4), (H5,H5)`
* Then continue hashing upward.

---

## ✅ Inclusion Proof (SPV)

Suppose you want to prove that **tx3** is inside the block.
You don’t need all transactions—just a **Merkle branch** (the sibling hashes along the path).

1. Provide `H3` (tx3’s hash).
2. Provide sibling `H4`.
3. Provide sibling `P12`.
4. Recompute upward:

   * `P34 = Hash(H3 || H4)`
   * `Root = Hash(P12 || P34)`
5. If your recomputed root matches the block header’s root, then **tx3 is included**.

---

## ⚡ Key Points

* **Efficiency:** Only one hash (Merkle root) commits to all transactions.
* **Security:** Changing any transaction changes its txid → changes the Merkle root.
* **Light clients:** Can verify inclusion without downloading the full block (SPV = Simplified Payment Verification).
* **Bitcoin:** Uses double-SHA-256 for hashing; Ethereum uses Keccak-256 and a more complex structure called a **Merkle Patricia Trie**.

# 🔑 Hash and Hashing

---

<img src="https://miro.medium.com/v2/resize:fit:1400/1*8hAoOurJFBfA-w1wKUAqIA.gif" width="450" alt="">

## 1. What is a Hash?

A **hash** is a fixed-length output (like a digital fingerprint) that is generated from any input data.

* Input can be anything: text, numbers, files, or images.
* Output is always the same length, no matter how big or small the input is.

👉 Example:

* Input: “Hello”
* Hash (using SHA-256):

  ```
  185f8db32271fe25f561a6fc938b2e264306ec304eda518007d1764826381969
  ```

### 📌 Key Properties

1. **Deterministic**: Same input → same output always.
2. **Fixed length**: Output size is fixed (e.g., SHA-256 always gives 256 bits).
3. **Irreversible**: You cannot find the original input from the hash.
4. **Collision resistance**: Very hard to find two different inputs with the same hash.
5. **Avalanche effect**: Small change in input → huge change in output.

---

## 2. What is Hashing?

**Hashing** is the process of applying a hash function to some data in order to get its hash value (digest).

It is like passing data through a **machine**:

```
Input → [Hash Function] → Hash Output
```

### Why do we need hashing?

* ✅ To check **data integrity** (has the data been changed?).
* ✅ To store **passwords** securely.
* ✅ To secure **blockchains and digital signatures**.
* ✅ To make **searching faster** in data structures like hash tables.

---

## 3. How Hashing Works (Step by Step)

1. **Take input** (text, file, etc.).
2. **Convert** it to binary (0s and 1s).
3. **Pad** the binary message so it fits the block size (e.g., SHA-256 uses 512-bit blocks).
4. **Initialize constants** (starting values defined by the algorithm).
5. **Process the message** in rounds (scrambling using AND, OR, XOR, shifts, etc.).
6. **Combine final results** to get the fixed-length hash output.

---

## 4. Example of Hashing (with SHA-256)

Let’s hash the word **HELLO**.

### Step 1. Convert input to binary

* H → 72 → `01001000`
* E → 69 → `01000101`
* L → 76 → `01001100`
* L → 76 → `01001100`
* O → 79 → `01001111`

Binary of HELLO =

```
01001000 01000101 01001100 01001100 01001111
```

---

### Step 2. Pad the message

* SHA-256 needs 512 bits.
* Add a `1` at the end, then enough `0`s, and finally the message length.

---

### Step 3. Initialize constants

* SHA-256 starts with 8 fixed numbers (predefined by the standard).

---

### Step 4. Process the message

* Message is split into chunks.
* For 64 rounds, the chunks are mixed with constants using simple operations like:

  * AND (both must be 1)
  * XOR (different bits → 1, same bits → 0)
  * Bit rotations (moving bits around).

---

### Step 5. Final Hash

After all rounds, we get the unique output:

```
185f8db32271fe25f561a6fc938b2e264306ec304eda518007d1764826381969
```

👉 Note: If we change just one letter (“HELLO” → “hello”), the hash becomes completely different.

---

## 5. Popular Hash Functions

### 🔸 MD5

* Output: 128 bits.
* Old and broken (collisions found).
* Only used for file checks, not for security.

### 🔸 SHA-1

* Output: 160 bits.
* Also broken and not secure.
* Used in old systems.

### 🔸 SHA-2 (SHA-224, SHA-256, SHA-384, SHA-512)

* Secure and widely used today.
* SHA-256 is used in Bitcoin, TLS, file integrity checks.

### 🔸 SHA-3 (Keccak)

* New design (sponge structure).
* Secure and flexible (can give variable output length).

### 🔸 RIPEMD-160

* 160-bit hash.
* Still used in Bitcoin addresses (combined with SHA-256).

### 🔸 BLAKE2 / BLAKE3

* Faster than SHA-2 and SHA-3.
* Secure and modern.

---

## 6. Password Hashing

Normal hashing (like SHA-256) is too fast, so attackers can guess millions of passwords quickly.
For passwords, we use **special slow hash functions**:

* **PBKDF2** → uses repeated hashing.
* **bcrypt** → slower, includes salt.
* **scrypt** → adds memory usage to slow down attackers.
* **Argon2** → best modern choice (used in new systems).

👉 All these also add **salt** (random extra data) so that two users with the same password don’t get the same hash.

---

## 7. Real-World Applications of Hashing

1. **File Integrity** 🛡️ – When you download a file, websites provide a hash. You can hash your file and compare both.
2. **Passwords** 🔐 – Websites store hashed passwords, not plain ones.
3. **Digital Signatures** ✍️ – A document is hashed and the hash is signed, not the whole document.
4. **Blockchain** ⛓️ – Each block references the hash of the previous block, ensuring immutability.
5. **Merkle Trees** 🌳 – Used in blockchains to verify many transactions quickly.
6. **Databases** 📂 – Hash tables allow very fast searching.

---

## 8. Summary

* A **hash** is a unique fixed-size digital fingerprint of data.
* **Hashing** is the process of generating it.
* Different algorithms exist (MD5, SHA-1, SHA-2, SHA-3, BLAKE).
* Passwords require special slow hashing (bcrypt, scrypt, Argon2).
* Real-world uses include security, blockchain, data integrity, and fast searching.
- Blockchain is maintained by different types of **nodes**.  
- It uses a **P2P gossip network** for fast, decentralized communication.  
- Transactions wait in the **mempool** until they are confirmed into blocks.

### 3️⃣ Block Lifecycle  

A blockchain block goes through several steps before it becomes part of the **immutable ledger**.  

---

## 🔄 Step 1: Transaction Creation & Signing  
- A user creates a **transaction** (e.g., Alice sends 5 BTC to Bob).  
- The transaction is **digitally signed** using the sender’s **private key**.  
- This ensures **authenticity** (only Alice can spend her coins).  

✍️ **Key Idea:** Digital signatures = proof of ownership.  

---

## 📡 Step 2: Transaction Broadcast to Network  
- The signed transaction is sent to the **peer-to-peer network**.  
- Nearby nodes receive it and spread it using the **gossip protocol**.  

📢 Within seconds, thousands of nodes know about the transaction.  

---

## ✅ Step 3: Validation by Nodes  
- Each node checks:  
  - Signature validity ✍️  
  - Double-spending attempt ❌  
  - Sufficient balance 💰  
- Invalid transactions are **rejected**.  
- Valid ones stay in the **mempool**.  

---

## 📦 Step 4: Block Proposal (Candidate Block)  
- Miners (PoW) or Validators (PoS) collect valid transactions.  
- They group them into a **candidate block**.  
- The block also contains:  
  - Block header (previous hash, timestamp, Merkle root).  
  - Nonce or validator signature.  

---

## 🌍 Step 5: Block Propagation  
- The candidate block is broadcast to the network.  
- Other nodes independently **verify**:  
  - All transactions inside are valid.  
  - Block hash / signature is correct.  
- If valid → nodes accept the block into their mempool/blockchain.  

---

## ⚖️ Step 6: Consensus & Finalization  
- Nodes apply a **consensus mechanism** (PoW, PoS, PBFT, etc.).  
- The block is officially **added to the chain**.  
- Once several confirmations pass → the block is considered **final**.  

---

## 🧠 Quick Example: Bitcoin  
1. Alice → Bob (signed tx)  
2. Tx enters mempool → miners see it  
3. Miner validates → includes it in candidate block  
4. Solves PoW puzzle (nonce)  
5. Block broadcast → other nodes verify  
6. Consensus → block appended to chain  

---

## 🔑 Summary  
- Transactions must be **signed, validated, and confirmed**.  
- A **block lifecycle** ensures:  
  - Authenticity ✅  
  - Security 🔒  
  - Agreement among all nodes 🌍

### 1️⃣ Why Consensus is Needed  

---

## 🤔 The Problem  

- In a blockchain network, **nodes cannot fully trust each other**.  
- Some nodes may:  
  - Fail 🛑 (go offline or crash)  
  - Act maliciously 😈 (alter transactions, inject false data)  
  - Behave selfishly 💰 (try to gain more rewards unfairly)  

Without consensus, different nodes might have **different versions of the ledger**, causing **chaos**.  

---

## 🎯 The Goal  

- Ensure that **all honest nodes** agree on:  
  - Which transactions are valid  
  - The **same order** of transactions  
  - A **single version of truth** (the blockchain)  

👉 This makes the system **trustless** → no central authority needed, but still secure.  

---

## 💸 Double-Spending Problem  

<img src="https://blocktrade.com/wp-content/uploads/2021/08/2021-25-08_700x508px_Double_Spending.jpg" width="350" alt="">

- A unique issue with digital money:  
  - Copying digital files is easy (like music or images).  
  - Without consensus, someone could **spend the same coin twice**.  

### Example:  
1. Alice sends **1 BTC** to Bob.  
2. At the same time, Alice sends **the same 1 BTC** to Carol.  
3. If nodes disagree, both could accept the transactions → **fraud!**  

👉 Consensus ensures **only one transaction** is valid.  

---

## 👥 Honest Majority Assumption  

- Most consensus algorithms assume:  
  - **>50% of nodes** (or economic weight in PoS) are **honest**.  
- If the **majority follows the rules**:  
  - Attackers cannot override the system.  
  - The ledger stays consistent and secure.  

📌 Example:  
- In **Proof of Work (PoW)** → majority of computational power must be honest.  
- In **Proof of Stake (PoS)** → majority of stake must be honest.  

---

## 🔑 Summary  

- Blockchain works in an **untrusted environment**.  
- **Consensus** makes sure:  
  - Honest nodes agree on one version of history.  
  - Double-spending is prevented.  
  - Security relies on the **honest majority assumption**.

### 2️⃣ Proof of Work (PoW)  

---

## ⚙️ Process  

- In PoW, miners compete to **solve a computational puzzle**.  
- The puzzle: find a **nonce** (random number) such that:  
- SHA-256(Block Data + Nonce) < Target

- The **target** defines difficulty → lower target = harder puzzle.  
- First miner to find a valid solution:  
  - Broadcasts the block.  
  - Other nodes verify the solution quickly.  
  - Block is added to blockchain.  

📌 **Key Idea:** Hard to find 🔍, easy to check ✅.  

---

## ⏱️ Difficulty Adjustment  

- To maintain stability, the puzzle difficulty **adjusts periodically**.  
- Example: **Bitcoin**  
  - Block time target = ~10 minutes.  
  - Every **2016 blocks (~2 weeks)**, the difficulty is recalculated.  
  - If blocks were mined too fast → difficulty increases.  
  - If too slow → difficulty decreases.  

---

## 💰 Incentives  

- Miners are rewarded for securing the network:  
  - **Block reward** (new coins minted).  
  - **Transaction fees** (from included transactions).  

- Example: Bitcoin block reward started at **50 BTC**, halves every ~4 years (Halving event).  

---

### 🔒 Why miners are rewarded

* In a blockchain, **miners** run powerful computers to solve cryptographic puzzles (Proof of Work).
* By doing so, they:

  * Validate transactions.
  * Add new blocks to the blockchain.
  * Secure the network against attacks (e.g., double-spending).

Without rewards, miners would have little reason to spend electricity and resources on this work.

---

### 💰 What is a block reward?

* **Block reward = New coins minted + (Transaction fees).**
* Each time a miner successfully adds a block:

  * The system **creates new coins out of nothing** (minting) and gives them to the miner.
  * Plus, the miner collects **fees from all transactions** inside that block.

This is called the **block reward**.

---

### 📉 Example – Bitcoin

* In Bitcoin, the block reward started at **50 BTC** in 2009.
* It halves roughly every 4 years (“halving events”):

  * 50 → 25 → 12.5 → 6.25 → now 3.125 BTC (since April 2024).
* Eventually, no new coins will be minted (around year 2140). After that, miners will rely only on **transaction fees**.

---

## ✅ Strengths of PoW  

- **Highly secure** 🔒 — attacking requires majority hash power.  
- **Battle-tested** ⚔️ — secures Bitcoin since 2009.  
- **Simple design** — straightforward to implement & verify.  

---

## ⚠️ Weaknesses of PoW  

- **Energy consumption** 🔋 — massive electricity required for mining.  
- **Limited scalability** 📉 — low transactions per second (Bitcoin ≈ 7 TPS).  
- **Centralization risk** — mining pools dominate due to hardware costs.  

---

## 🪙 Example: Bitcoin Mining  

- Miners collect transactions from mempool.  
- Build a candidate block.  
- Continuously change nonce → calculate hash.  
- When hash < target → solution found.  
- Block is broadcast → verified by nodes.  
- Miner receives **block reward + fees**.  

---

## 🔑 Summary  

- **PoW = “security through computation.”**  
- Ensures fairness & prevents double-spending.  
- Secures the largest cryptocurrencies like **Bitcoin**.

### 3️⃣ Proof of Stake (PoS)  

---

## ⚙️ Process  

- In **Proof of Stake**, there are **no miners** burning electricity.  
- Instead, **validators** are chosen to propose & verify blocks **based on their stake**.  
- Stake = tokens **locked (staked)** in the protocol.  
- The more tokens staked → the higher the chance of being chosen as block proposer.  

📌 **Key Idea:** “Skin in the game” → validators risk their own funds.  

---

## 💰 Rewards & Penalties  

- **Rewards:**  
  - Validators earn new coins + transaction fees for proposing/validating honestly.  

- **Penalties (Slashing):**  
  - If a validator cheats (double-signs, invalid blocks, offline too often),  
    → part of their stake is **slashed (burned/lost)**.  
  - This discourages malicious behavior.  

---

## ✅ Advantages of PoS  

1. **Energy-efficient** 🔋  
   - No need for massive mining hardware.  
   - ~99% less energy than PoW.  

2. **Faster finality** ⚡  
   - Blocks are confirmed in seconds (not minutes).  

3. **Scalability potential** 🌍  
   - Supports higher transaction throughput.  

---

## ⚠️ Weaknesses of PoS  

1. **Wealth concentration** 💰  
   - Rich validators may dominate → “the rich get richer” problem.  

2. **Long-range attacks** ⏳  
   - Attackers may try to rewrite history from old checkpoints.  
   - Mitigated by **checkpointing & slashing**.  

3. **Nothing-at-stake problem** ❌  
   - Validators might try to support multiple forks (since no cost like PoW).  
   - Solved with slashing conditions.  

---

## 🪙 Example: Ethereum 2.0  

- Ethereum transitioned from PoW → PoS in **2022 (The Merge)**.  
- Uses **Casper (Friendly Finality Gadget)** and the **Beacon Chain** for coordination.  
- Validators stake **32 ETH** to participate.  
- Network now consumes **~99.95% less energy** vs PoW Ethereum.  

---

## 🔑 Summary  

- **PoS = “security through economic stake.”**  
- Validators risk their money → dishonest behavior punished.  
- Modern blockchains like **Ethereum 2.0, Cardano, Polkadot** use PoS.

### 4️⃣ Practical Byzantine Fault Tolerance (PBFT)  

---

## ⚔️ The Problem: Byzantine Generals’ Dilemma  

- In distributed systems, nodes may:  
  - Fail 🛑 (go offline)  
  - Act maliciously 😈 (send false information)  
- This is known as the **Byzantine Generals’ Problem**:  
  - How can multiple generals (nodes) agree on a single plan (ledger state)  
    when some may be traitors (malicious)?  

📌 PBFT is designed to solve this problem in **small-to-medium permissioned networks**.  

---

## ⚙️ The Process (Multi-Phase Voting)  

PBFT works through **three main phases of voting**:  

1. **Pre-Prepare** 🟢  
   - A leader (primary node) proposes a block.  

2. **Prepare** 🟡  
   - Other nodes broadcast that they received the proposal.  
   - Nodes verify validity of the block.  

3. **Commit** 🔵  
   - Nodes confirm final agreement.  
   - If ≥ 2/3 majority agree → block is added.  

💡 **Tolerates up to f malicious nodes in a network of 3f + 1 nodes.**  

---

## 🏢 Use Cases  

- **Enterprise / Permissioned blockchains** where participants are known & trusted to some degree:  
  - Hyperledger Fabric  
  - Tendermint (used in Cosmos ecosystem)  

📌 Ideal for **business networks, banking consortia, supply chain systems**.  

---

## ✅ Advantages  

- **Low latency** ⚡ → blocks confirmed quickly.  
- **Deterministic finality** 🔒 → once block is committed, it cannot be reversed.  
- **Energy efficient** 💡 → no mining required.  

---

## ⚠️ Limitations  

- **Not scalable to thousands of nodes** 🚫  
  - Communication overhead grows **O(n²)** (each node talks to all others).  
- Works best for **small-to-medium networks** (10s or 100s of nodes).  

---

## 🔑 Summary  

- **PBFT = consensus by voting** 🗳️  
- Tolerates malicious actors up to 1/3 of nodes.  
- Great for **enterprise / permissioned systems** (fast + final).  
- Limited scalability makes it unsuitable for large public blockchains.

### 6️⃣ Digital Signatures & Keys (Deep Dive)  

---

## 🔑 Public/Private Key Pairs  

- Every blockchain user has a **key pair**:  
  - **Private Key 🔒** → Secret, known only to the owner.  
  - **Public Key 🔑** → Shared openly with others.  

- **Mathematical relationship:**  
  - Private key → used to **sign** transactions.  
  - Public key → used to **verify** transactions.  

💡 **Analogy:**  
- Private key = your handwritten signature ✍️  
- Public key = handwriting reference anyone can check.  

---

## 💰 Transaction Security  

- Only the **private key holder** can authorize spending.  
- Example:  
  - Alice sends 1 BTC to Bob → she signs the transaction with her private key.  
  - All nodes can verify it using Alice’s public key.  

📌 **Result:** No one else can spend Alice’s coins.  

---

## 🔢 Digital Signature Algorithms in Blockchain  

1. **ECDSA (Elliptic Curve Digital Signature Algorithm)**  
   - Used in **Bitcoin, Ethereum (pre-merge)**.  
   - Strong cryptography based on elliptic curves.  

2. **EdDSA (Edwards-Curve Digital Signature Algorithm)**  
   - Used in **Cardano, Zcash, and many newer blockchains**.  
   - More efficient & secure → faster verification.  

---

## 🛡️ Verification Process  

1. User signs a message with their **private key**.  
2. Signature + message are sent to the network.  
3. Other nodes use the **public key** to verify:  
   - ✅ Message was signed by the rightful owner.  
   - ✅ Message has not been altered (integrity).  
   - ✅ Authentic → prevents forgery.  

---

## 📦 Example Flow:  

1. Alice → creates transaction “Send 5 coins to Bob.”  
2. Signs with her **private key**.  
3. Transaction broadcast to network.  
4. Nodes use Alice’s **public key** to verify signature.  
5. If valid → transaction accepted into mempool.  

---

## 🔑 Summary  

- Digital signatures guarantee **ownership + authenticity** in blockchain.  
- **Private key = authority** to spend coins.  
- **Public key = tool** to verify transactions.  
- Algorithms: **ECDSA** (Bitcoin) & **EdDSA** (newer chains).

### 7️⃣ Wallets & Key Infrastructure  

---

## 🔥 Hot Wallets  

- **Definition:** Wallets connected to the **internet**.  
- Examples: **MetaMask, Trust Wallet, mobile crypto apps**.  
- ✅ Pros:  
  - Easy to set up  
  - Convenient for daily use (trading, payments)  
  - Fast access to funds  
- ❌ Cons:  
  - Higher risk of **hacking, phishing, malware**  
  - Dependent on device security  

---

## ❄️ Cold Wallets  

- **Definition:** Wallets stored **offline**, not directly connected to internet.  
- Examples: **Hardware wallets (Ledger, Trezor), paper wallets**.  
- ✅ Pros:  
  - Very secure → immune to online hacks  
  - Ideal for **long-term storage** of large funds  
- ❌ Cons:  
  - Less convenient for frequent transactions  
  - Risk of physical loss (if paper/hardware is stolen or damaged)  

---

## 👥 Custodial vs Non-Custodial  

1. **Custodial Wallets** 🏦  
   - A **third party (bank, exchange, platform)** controls your private keys.  
   - ✅ Easy to recover password / reset access.  
   - ❌ “Not your keys, not your coins.” Risk of exchange hacks.  

2. **Non-Custodial Wallets** 🔑  
   - You **control the private keys** directly.  
   - ✅ Full control & independence.  
   - ❌ If you lose your key → funds are lost forever.  

---

## 🛡️ Key Safety  

- **Seed Phrase:**  
  - A 12–24 word recovery phrase generated by wallet.  
  - Backup this phrase securely (offline, written on paper, stored safely).  

- **Never share private keys** 🚫  
  - Anyone with your private key has full access to your funds.  

- **Best Practices:**  
  - Use **cold storage** for large funds.  
  - Keep **multiple backups** of seed phrases.  
  - Enable **2FA (Two-Factor Authentication)** where possible.  

---

## 🔑 Summary  

- **Hot wallets** = convenience but less secure.  
- **Cold wallets** = maximum security for long-term funds.  
- **Custodial wallets** = someone else manages your keys.  
- **Non-custodial wallets** = you manage your own keys.  
- **Golden Rule:** If you don’t control your keys → you don’t control your coins.

- Enterprise blockchain platform with **permissioned network**.  
- Modular architecture (plug-and-play components).  
- Supports **smart contracts** (called **chaincode**).  
- Strong focus on **privacy** and **scalability**.  

---

### 3️⃣ Hyperledger Fabric Architecture  

- **Membership Service Provider (MSP):** Manages identity of participants.  
- **Peers (nodes):**  
  - Endorsing peers → simulate & endorse transactions.  
  - Committing peers → validate and commit blocks.  
- **Ordering Service:**  
  - Defines transaction order.  
  - Ensures consistency across network.  
- **Channels:**  
  - Private communication subnets between participants.  
  - Enable confidential transactions.  
- **Chaincode (Smart Contracts):**  
  - Business logic implemented in Go/Java/Node.js.  
  - Runs inside Docker containers.  

---

### 4️⃣ Hyperledger Fabric Transaction Flow  

1. **Proposal:** Client sends transaction proposal to endorsing peers.  
2. **Endorsement:** Peers simulate & sign results (without committing).  
3. **Ordering:** Ordering service creates transaction order.  
4. **Validation & Commit:** Peers validate endorsements → commit block.  

---

- **Permissioned network** (identity-based access).  
- **Pluggable consensus** (Kafka, Raft, BFT options).  
- **Channels for privacy** (separate ledgers per group).  
- **High throughput** (hundreds of TPS possible).

## 🔹 A. Setup

































````python
# We need some basic Python libraries
import time, json, hashlib, secrets

# Helper function: SHA-256 hashing
def sha256_hex(s: str) -> str:
    """
    Takes a string, encodes it as UTF-8,
    and returns its SHA-256 hash as a hex string.
    """
    return hashlib.sha256(s.encode("utf-8")).hexdigest()
````

---

## 🔹 B. Proof of Work (Mining a Block)



````python

# Function to "mine" a block by finding a nonce
def mine_block(data: str, prev_hash: str, difficulty: int):
    """
    - data: transaction data
    - prev_hash: hash of the previous block
    - difficulty: number of leading zeros required in hash
    """
    prefix = "0" * difficulty  # e.g. "0000" if difficulty=4
    nonce = 0                  # start from 0, try upward
    start = time.time()

    while True:
        # Block contents = data + prev_hash + nonce
        block_string = f"{data}{prev_hash}{nonce}"
        block_hash = sha256_hex(block_string)

        # Check if hash meets the difficulty target
        if block_hash.startswith(prefix):
            end = time.time()
            print("✅ Block mined!")
            print("Data       :", data)
            print("Nonce      :", nonce)
            print("Hash       :", block_hash)
            print("Time taken :", round(end-start, 3), "seconds")
            break

        # Try the next nonce
        nonce += 1

# Example: mine a block with difficulty 4
mine_block("Alice pays Bob 5 coins", "0"*64, 4)
````

---

## 🔹 C. Difficulty vs Time



````python
# Let’s see how difficulty affects mining time
for diff in [3, 4, 5]:
    print(f"\n--- Mining with difficulty {diff} ---")
    mine_block("Test Tx", "0"*64, diff)
````

👉 Students will notice that **higher difficulty → longer mining time**.

## 🔹 D. Experiment: Tampering with a Block



````python

# Step 1: Mine a valid block
data = "Carol pays Dave 3 coins"
prev_hash = "0"*64
difficulty = 4

print("⛏️ Mining original block...")
block_string = f"{data}{prev_hash}0"
block_hash = sha256_hex(block_string)
print("Original Hash:", block_hash)

# Step 2: Tamper with data
tampered_data = "Carol pays Dave 300 coins (tampered!)"
tampered_string = f"{tampered_data}{prev_hash}0"
tampered_hash = sha256_hex(tampered_string)

print("\n🚨 After tampering:")
print("Tampered Hash :", tampered_hash)
print("Valid?", tampered_hash.startswith("0"*difficulty))
````

👉 **Key lesson:** Changing even **1 character** makes the hash invalid → must **re-mine**.

## 🔹 E. Cryptographic Demo (Keys & Signatures)



````python
# Install ECDSA library for digital signatures
!pip -q install ecdsa
from ecdsa import SigningKey, SECP256k1
````

````python
# 1) Generate private & public keys
sk = SigningKey.generate(curve=SECP256k1)  # Private key
vk = sk.get_verifying_key()                # Public key

print("🔒 Private Key (keep secret!):", sk.to_string().hex()[:32], "... (truncated)")
print("🔑 Public Key (share safely):", vk.to_string().hex()[:40], "... (truncated)")
````

````python
# 2) Derive a simple pseudo-address (like BTC/ETH style)
def pseudo_address(pubkey_bytes: bytes) -> str:
    """Simplified: hash the public key into a shorter address (demo only)."""
    sha = hashlib.sha256(pubkey_bytes).hexdigest()
    return sha[-40:]  # take last 40 hex chars (like Ethereum address)

address = pseudo_address(vk.to_string())
print("🏦 Pseudo Wallet Address:", "0x"+address)
````

````python
# 3) Sign & Verify a message
message = "Alice sends 10 tokens to Bob"
message_bytes = message.encode()

# Sign with PRIVATE key
signature = sk.sign(message_bytes)
print("✍️ Signature:", signature.hex()[:32], "... (truncated)")

# Verify with PUBLIC key
valid = vk.verify(signature, message_bytes)
print("✅ Is the signature valid?", valid)

# Try tampering
tampered_msg = "Alice sends 1000 tokens to Bob".encode()
try:
    vk.verify(signature, tampered_msg)
except:
    print("🚨 Tampered message detected: Verification failed")
````

* Other groups **check the block math**.
* If valid → everyone accepts it and uses that Block Hash as **PrevHash** for the next block.
* If two groups mine at the same time → keep both versions until the next block is mined.
* **Rule: The longest valid chain wins.**

---

### **6 – Race Goal**

* Keep mining 3–4 rounds.
* **Winner = the group that builds the longest valid chain.**
