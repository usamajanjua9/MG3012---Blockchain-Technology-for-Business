#### 📖 What is a Smart Contract?

🧾 A **smart contract** is a **self-executing digital agreement** with the terms of the contract directly written into code. It is stored on a blockchain (e.g., Ethereum) and **automatically runs** when predefined conditions are met.

🧠 Think of it as a **"trustless digital vending machine"**:

* Insert money (e.g., ETH)
* Select item (e.g., function call)
* The smart contract delivers (e.g., digital asset) **without human intervention**

#### 🧱 Key Features:

* ✅ **Immutable**: Once deployed, the contract **cannot be changed**
* 🔍 **Transparent**: Code is visible to anyone on the blockchain
* 🕹 **Autonomous**: Executes logic **automatically**
* 🤝 **Trustless**: No need to trust a third party — **code is law**
* 💸 **Handles digital assets**: Can send/receive ETH, tokens

### 🔹 **What is Solidity?**

**Definition:**
Solidity is a **high-level, contract-oriented, statically typed** programming language used for writing smart contracts on **Ethereum** and other **EVM-compatible** blockchains.

**Key Features:**

* **Contract-oriented**: Organizes code into smart contracts (similar to classes).
* **Statically typed**: Data types must be declared.
* Compiles to **EVM bytecode**.
* Influenced by **JavaScript, Python, and C++**.

**Example Smart Contract:**

```solidity
pragma solidity ^0.8.0;

contract HelloWorld {
    string public message = "Hello, Ethereum!";
}
```

---

### 🔹 **✅ Pragma and Versioning**

**Definition:**
`pragma` is a directive that tells the compiler which version of Solidity to use.

**Why It's Important:**
Ensures compatibility. Newer versions might break older code.

**Syntax:**

```solidity
pragma solidity ^0.8.0;
```

**Explanation:**

* `^0.8.0` allows any version from `0.8.0` to less than `0.9.0`.
* To restrict exactly: `pragma solidity 0.8.20;`
* To specify a range: `pragma solidity >=0.7.0 <0.9.0;`

✅ **Best Practice:** Always specify the version range to avoid future errors.

---

### 🔹 **✅ Structure of a Smart Contract**

**Basic Structure:**

```solidity
pragma solidity ^0.8.0;

contract MyContract {
    // State Variables
    uint public myNumber;

    // Constructor
    constructor() {
        myNumber = 10;
    }

    // Functions
    function setNumber(uint _num) public {
        myNumber = _num;
    }

    function getNumber() public view returns (uint) {
        return myNumber;
    }
}
```

**Explanation:**

* `contract`: Defines the smart contract (like a class).
* **State Variable**: `myNumber` is stored on the blockchain.
* **Constructor**: Initializes values when contract is deployed.
* **Function**: Allows interaction with contract.

✅ **Tip:** A contract is like a container that holds all your logic, data, and rules.

# 🔹 **State Variables vs Local Variables in Solidity**

---

## 1️⃣ **📘 What is a State Variable?**

### ✅ Definition:

> A **state variable** is a variable declared **outside any function**, and its value is **stored permanently** on the blockchain.

### 🧠 Explanation:

* Lives **as long as the contract exists**.
* Used to **store important data** like balances, totals, names, etc.
* Costs **gas** when you write/update them.
* Can be made **public** to automatically create a getter function.

### 🏠 Real-world Analogy:

> Imagine a **whiteboard in a classroom**. Whatever is written on it stays even after the class ends. Everyone can see it unless erased.

### 🧪 Example Code:

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract StateExample {

    uint public number = 5; // 🟢 State variable (stored on blockchain)

    function getNumber() public view returns (uint) {
        return number; // 🔄 Read state variable (no gas cost)
    }

    function setNumber(uint _num) public {
        number = _num; // 🛠️ Write to state variable (costs gas)
    }
}
```

---

## 2️⃣ **📘 What is a Local Variable?**

### ✅ Definition:

> A **local variable** is declared **inside a function** and exists **only while the function runs**.

### 🧠 Explanation:

* **Temporary**: Disappears after function finishes.
* **Does not cost gas to store**, only computing cost.
* Used for **intermediate calculations**, checks, etc.
* **Cannot be accessed** outside the function.

### 🛠 Real-world Analogy:

> Think of **writing something on a piece of paper during class** and throwing it away after. It’s **temporary** and **only you can see it**.

### 🧪 Example Code:

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract LocalExample {

    function calculateDouble(uint x) public pure returns (uint) {
        uint result = x * 2;  // 🔵 Local variable
        return result;        // Returned value, but 'result' disappears after
    }
}
```

---

## 🔄 Putting Both Together (Comparison Example)

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract VariableTypes {

    // 🟢 State Variable: Stored permanently on blockchain
    uint public count = 10;

    function updateCount(uint _input) public {
        // 🔵 Local Variable: Used temporarily inside function
        uint newCount = _input + 5;

        // Assign local variable to state variable
        count = newCount;
    }
}
```

---

## 📊 Comparison Table

| Feature          | 🟢 State Variable           | 🔵 Local Variable               |
| ---------------- | --------------------------- | ------------------------------- |
| Declared         | Outside functions           | Inside functions                |
| Lifetime         | Until contract is destroyed | Only during function execution  |
| Storage Location | Blockchain (Storage)        | Memory (Temporary)              |
| Gas Cost         | Writing costs gas 💸        | No storage gas cost 💡          |
| Accessibility    | Available to all functions  | Only available in that function |
| Example          | `uint public value;`        | `uint temp = x + 1;`            |

---

## 🧠 Summary to Teach Students:

| ✅ Key Point       | 💬 What to Tell Students                                    |
| ----------------- | ----------------------------------------------------------- |
| State Variable    | Used to **store data** on the blockchain **permanently**    |
| Local Variable    | Used **inside functions** and **disappears** after          |
| Gas Cost          | Writing to state variable costs **gas**, local does **not** |
| Real-life Example | Whiteboard (state) vs Notebook page (local)                 |

## 🔢 **Basic Data Types & Keywords in Solidity**

Solidity provides a rich set of **data types** that are similar to those in other programming languages but are specially optimized for blockchain usage.

Each data type stores different kinds of information, and choosing the correct one is essential for performance, security, and gas efficiency.

---

### ✅ **1. Integer Types (`uint`, `int`)**

#### 📌 `uint` → **Unsigned Integer**

* Stores only **positive numbers**, including zero.
* Commonly used to store counters, IDs, balances, etc.

```solidity
uint256 public totalSupply = 1000;
```

* You can also specify bit size: `uint8`, `uint16`, ..., up to `uint256`.
* Default: `uint` = `uint256`.

#### 📌 `int` → **Signed Integer**

* Can store **both positive and negative** numbers.

```solidity
int256 public temperature = -20;
```

**⚠️ Note:** Overflow/underflow is checked automatically in Solidity >= 0.8.0

---

### ✅ **2. Boolean Type (`bool`)**

Used to store true/false values.

```solidity
bool public isActive = true;

function deactivate() public {
    isActive = false;
}
```

✅ Useful in toggles, condition checks, access control, etc.

---

### ✅ **3. String Type (`string`)**

Used to store text.

```solidity
string public name = "Ethereum";
```

✅ Strings are **expensive in gas**, so avoid frequent storage unless necessary.

---

### ✅ **4. Address Type (`address`)**

Stores Ethereum wallet addresses (20 bytes, 160 bits).

```solidity
address public owner;

constructor() {
    owner = msg.sender; // Sets the owner as the deployer
}
```

You can also:

* Check balance: `owner.balance`
* Transfer Ether: `payable(owner).transfer(1 ether);`

---

### ✅ **5. Bytes Type (`bytes1` to `bytes32`, `bytes`)**

* **Fixed-size** binary data: `bytes1`, `bytes2`, ..., up to `bytes32`.
* **Dynamically-sized** binary data: `bytes`.

Example:

```solidity
bytes32 public dataHash = keccak256(abi.encodePacked("hello"));
```

---

### ✅ **6. Constants and Immutables**

#### 📌 `constant` → value is fixed **at compile time**

```solidity
uint constant MAX_USERS = 100;
```

#### 📌 `immutable` → value is fixed **at deployment**

```solidity
address immutable DEPLOYER;

constructor() {
    DEPLOYER = msg.sender;
}
```

✅ **Why use them?**

* Gas-efficient
* Cannot be changed later → increases security

---

### ✅ Summary Table

| Type        | Description                  | Example                       |
| ----------- | ---------------------------- | ----------------------------- |
| `uint`      | Positive integers            | `uint public x = 5;`          |
| `int`       | Positive + negative integers | `int public temp = -10;`      |
| `bool`      | Boolean values               | `bool public isTrue = true;`  |
| `string`    | Text strings                 | `string public name = "BTC";` |
| `address`   | Ethereum wallet address      | `address public user;`        |
| `bytes32`   | Fixed-size byte data         | `bytes32 public hash;`        |
| `constant`  | Value fixed at compile time  | `uint constant MAX = 100;`    |
| `immutable` | Value fixed at deploy time   | `address immutable creator;`  |

---

🔎 **Tip for Beginners:**
If you're unsure which data type to use:

* Use `uint` for numbers unless you need negative values.
* Use `string` or `bytes` for text.
* Use `address` for wallets.
* Use `bool` for switches or decisions.

# ✅ **7. Operators in Solidity**

Operators in Solidity help you **perform actions** on variables, such as:

* Calculations
* Comparisons
* Logic checks
* Assignments

---

## 📘 1️⃣ Arithmetic Operators

Used to **perform math operations**.

| Operator | Description        | Example (`a=10, b=3`) | Result |
| -------- | ------------------ | --------------------- | ------ |
| `+`      | Addition           | `a + b`               | `13`   |
| `-`      | Subtraction        | `a - b`               | `7`    |
| `*`      | Multiplication     | `a * b`               | `30`   |
| `/`      | Division           | `a / b`               | `3`    |
| `%`      | Modulo (Remainder) | `a % b`               | `1`    |

### 🔢 Example:

```solidity
pragma solidity ^0.8.0;

contract ArithmeticDemo {
    function mathOps(uint a, uint b) public pure returns (uint, uint, uint, uint, uint) {
        return (
            a + b,  // Addition
            a - b,  // Subtraction
            a * b,  // Multiplication
            a / b,  // Division
            a % b   // Modulo
        );
    }
}
```

---

## 📗 2️⃣ Comparison Operators

Used to **compare values**. The result is always **true or false**.

| Operator | Description           | Example (`a=5, b=10`) | Result  |
| -------- | --------------------- | --------------------- | ------- |
| `==`     | Equal to              | `a == b`              | `false` |
| `!=`     | Not equal to          | `a != b`              | `true`  |
| `<`      | Less than             | `a < b`               | `true`  |
| `>`      | Greater than          | `a > b`               | `false` |
| `<=`     | Less than or equal    | `a <= b`              | `true`  |
| `>=`     | Greater than or equal | `a >= b`              | `false` |

### 🧪 Example:

```solidity
pragma solidity ^0.8.0;

contract ComparisonDemo {
    function compare(uint a, uint b) public pure returns (bool, bool, bool, bool, bool, bool) {
        return (
            a == b,
            a != b,
            a < b,
            a > b,
            a <= b,
            a >= b
        );
    }
}
```

---

## 📕 3️⃣ Logical Operators

Used to combine **true/false conditions**.

| Operator | Description | Example             | Result                       |          |   |           |                           |
| -------- | ----------- | ------------------- | ---------------------------- | -------- | - | --------- | ------------------------- |
| `&&`     | AND         | `(a > 5 && b < 10)` | `true` only if both are true |          |   |           |                           |
| \|\|   |             | --                  | OR                           | \`(a > 5 \|\| b < 10)\` | `true` if any one is true |
| `!`      | NOT         | `!(a > b)`          | `true` if condition is false |          |   |           |                           |

### 🧠 Example:

```solidity
pragma solidity ^0.8.0;

contract LogicalDemo {
    function logic(uint a, uint b) public pure returns (bool, bool, bool) {
        return (
            a > 5 && b < 10,   // AND
            a > 5 || b < 10,   // OR
            !(a > b)           // NOT
        );
    }
}
```

---

## 📙 4️⃣ Assignment Operators

Used to **store or update** values.

| Operator | Meaning             | Example  | Result (if `a=5`) |
| -------- | ------------------- | -------- | ----------------- |
| `=`      | Assign              | `a = 10` | `a = 10`          |
| `+=`     | Add and assign      | `a += 3` | `a = 8`           |
| `-=`     | Subtract and assign | `a -= 2` | `a = 3`           |
| `*=`     | Multiply and assign | `a *= 2` | `a = 10`          |
| `/=`     | Divide and assign   | `a /= 5` | `a = 2`           |

### ✏️ Example:

```solidity
pragma solidity ^0.8.0;

contract AssignmentDemo {
    function assignOps(uint a) public pure returns (uint, uint) {
        a += 5; // a = a + 5
        a *= 2; // a = a * 2
        return (a, a / 2);
    }
}
```

# ✅ Solidity Data Structures

---

## 🔹 1️⃣ Array

### ✅ Definition:

An **array** is a list of items stored in a specific order. Each item has an index (starts at 0).

### 🧠 Analogy:

Like a **row of boxes** labeled 0, 1, 2... You can store a number in each.

### 🔧 Code Example:

```solidity
pragma solidity ^0.8.0;

contract MyArray {
    uint[] public numbers;

    function addNumber(uint _num) public {
        numbers.push(_num); // Add to the end
    }

    function getNumber(uint index) public view returns (uint) {
        return numbers[index];
    }
}
```

### 📘 Use Case:

* Store test scores, balances, or list of users.

---

## 🔹 2️⃣ Struct

### ✅ Definition:

A **struct** is a custom data type that groups multiple values under one name.

### 🧠 Analogy:

Like a **form** with fields: Name, Age, Address.

### 🔧 Code Example:

```solidity
contract MyStruct {
    struct Student {
        string name;
        uint age;
    }

    Student public student;

    function setStudent(string memory _name, uint _age) public {
        student = Student(_name, _age);
    }

    function getStudent() public view returns (string memory, uint) {
        return (student.name, student.age);
    }
}
```

### 📘 Use Case:

* Store information like user profile, products, vehicles.

---

## 🔹 3️⃣ Mapping

### ✅ Definition:

A **mapping** stores data like a dictionary or phonebook: `key => value`.

### 🧠 Analogy:

Like a **contact list**: name → number.

### 🔧 Code Example:

```solidity
contract MyMapping {
    mapping(address => uint) public balances;

    function setBalance(uint _amount) public {
        balances[msg.sender] = _amount;
    }

    function getBalance(address user) public view returns (uint) {
        return balances[user];
    }
}
```

### 📘 Use Case:

* Store balances, roles, voting status.

---

## 🔹 4️⃣ Nested Mapping

### ✅ Definition:

A **mapping inside another mapping**, like `mapping(address => mapping(uint => uint))`.

### 🧠 Analogy:

Like a **year-wise diary for each person**.

### 🔧 Code Example:

```solidity
contract MyNestedMapping {
    mapping(address => mapping(uint => uint)) public scores;

    function setScore(uint year, uint score) public {
        scores[msg.sender][year] = score;
    }

    function getScore(address user, uint year) public view returns (uint) {
        return scores[user][year];
    }
}
```

### 📘 Use Case:

* Store data like (user → year → balance or rating).

---

## 🔹 5️⃣ Array of Structs

### ✅ Definition:

An array that stores multiple structured objects.

### 🧠 Analogy:

Like a **list of forms** filled by multiple students.

### 🔧 Code Example:

```solidity
contract MyArrayStruct {
    struct Task {
        string title;
        bool done;
    }

    Task[] public tasks;

    function addTask(string memory _title) public {
        tasks.push(Task(_title, false));
    }

    function getTask(uint index) public view returns (string memory, bool) {
        Task memory t = tasks[index];
        return (t.title, t.done);
    }
}
```

### 📘 Use Case:

* Todo apps, lists of items, users, or NFTs.

## 🔹 6️⃣ Mapping of Structs

### ✅ Definition:

A mapping that links a key to a struct object.

### 🧠 Analogy:

# 🔁 Loops in Solidity

Solidity supports **for**, **while**, and **do...while** loops, just like other languages.  
⚠️ Remember this issue: loops consume **gas**. If a loop runs too many times, the transaction may fail due to *out-of-gas*.  

---

## ✅ Example Contract with All Loops

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract LoopExamples {

    // -------------------------------
    // 1️⃣ For Loop Example
    // -------------------------------
    // Function to calculate sum of numbers from 1 to n
    function sumNumbers(uint n) public pure returns (uint) {
        uint sum = 0;
        for (uint i = 1; i <= n; i++) {
            sum += i;   // Add i to sum in each iteration
        }
        return sum;  // Output: total sum
    }

    // -------------------------------
    // 2️⃣ While Loop Example
    // -------------------------------
    // Function to calculate factorial of n
    function factorial(uint n) public pure returns (uint) {
        uint result = 1;
        uint i = n;
        while (i > 0) {
            result *= i; // Multiply result by current i
            i--;         // Decrease i
        }
        return result;   // Output: factorial of n
    }

    // -------------------------------
    // 3️⃣ Do...While Loop Example
    // -------------------------------
    // Function to count down from n, returns number of iterations
    function countDown(uint n) public pure returns (uint) {
        uint counter = 0;
        do {
            counter += 1;  // Increase counter
            n--;           // Decrease n
        } while (n > 0);
        return counter;    // Output: how many times loop ran
    }
}



---

## ⚙️ **Functions Overview in Solidity**

Functions are the **building blocks** of smart contracts. They define **what the contract can do**, like read data, write data, send Ether, or run logic.

---

### ✅ **1. Defining a Function**

Basic syntax:

```solidity
function functionName() public returns(type) {
    // logic here
}
```

# ✅ **2. Function Visibility Modifiers in Solidity**

Function visibility modifiers control **who is allowed to call a function** in a smart contract.

---

## 📊 Visibility Modifiers Table

| 🔐 Modifier | 👥 **Who Can Call the Function?**                                  |
| ----------- | ------------------------------------------------------------------ |
| `public`    | Anyone (inside or outside the contract, including other contracts) |
| `private`   | Only from **inside the same contract**                             |
| `internal`  | From this contract and **child contracts (inheritance)**           |
| `external`  | Only from **outside the contract** (not from internal functions)   |

---

## 🔍 Easy Real-World Analogy

| Modifier   | Real-Life Example                                                        |
| ---------- | ------------------------------------------------------------------------ |
| `public`   | A public park — **anyone** can enter                                     |
| `private`  | Your bedroom — **only you** can enter                                    |
| `internal` | Company staff room — **staff and their interns** can enter               |
| `external` | A public telephone — **only outsiders** can call in, not you from inside |

---

## 🧪 Examples (One by One)

---

### 1️⃣ `public` – **Anyone can call it**

* You can call it from inside the contract or outside (like from another contract or user).

```solidity
contract MyContract {
    // ✅ Public function
    function sayHello() public pure returns (string memory) {
        return "Hello, world!";
    }

    function callSayHello() public pure returns (string memory) {
        return sayHello(); // ✅ Can be called from inside
    }
}
```

⛳ **Used for most functions that interact with users**.

---

### 2️⃣ `private` – **Only this contract can call**

* Can’t be accessed outside this contract (not even by inherited contracts).

```solidity
contract MyContract {
    // 🔒 Private function
    function secretCode() private pure returns (uint) {
        return 12345;
    }

    function reveal() public pure returns (uint) {
        return secretCode(); // ✅ Allowed, called from inside
    }
}
```

❌ This function **cannot be accessed** from other contracts or externally.

---

### 3️⃣ `internal` – **This contract and its children can call**

* Used when you want to **share logic across contracts** using inheritance.

```solidity
contract Parent {
    // 🧬 Internal function
    function internalGreet() internal pure returns (string memory) {
        return "Hello from parent!";
    }
}

contract Child is Parent {
    function callParentGreeting() public pure returns (string memory) {
        return internalGreet(); // ✅ Allowed because of inheritance
    }
}
```

🧠 Think of it as **protected** access in object-oriented programming.

---

### 4️⃣ `external` – **Only outside the contract can call**

* Cannot be called from other functions inside the same contract directly (unless using `this.`).

```solidity
contract MyContract {
    // 🌐 External function
    function getName() external pure returns (string memory) {
        return "External Function";
    }

    // This will fail:
    /*
    function failCall() public pure returns (string memory) {
        return getName(); // ❌ ERROR: cannot call external function internally
    }
    */

    // ✅ This works:
    function properCall() public view returns (string memory) {
        return this.getName(); // ✅ call using 'this' pointer (less efficient)
    }
}
```

💡 Best for **APIs** or when you expect the function to be called by **users or other contracts only**.

| Modifier   | Can be Called From This Contract? | Can be Called From Outside? | Inherited Contracts? | Use Case            |
| ---------- | --------------------------------- | --------------------------- | -------------------- | ------------------- |
| `public`   | ✅ Yes                             | ✅ Yes                       | ✅ Yes                | User functions      |
| `private`  | ✅ Yes                             | ❌ No                        | ❌ No                 | Internal logic only |
| `internal` | ✅ Yes                             | ❌ No                        | ✅ Yes                | Shared logic        |
| `external` | ❌ No (except via `this`)          | ✅ Yes                       | ❌ No                 | APIs                |

* Only uses inputs and local variables.

### 🔍 Example:

```solidity
pragma solidity ^0.8.0;

contract PureExample {
    
    // ✅ Does not use any state variables
    function add(uint a, uint b) public pure returns (uint) {
        return a + b;
    }
}
```

### 📌 Real-world analogy:

## 🔹 5. **Default Functions in Solidity: `fallback()` and `receive()`**

These are **special functions** that run automatically when something happens with **Ether or unknown calls**.

---

### 📩 `receive()`

* Runs when **plain Ether** is sent (no data, no function call).
* Must include `payable`.

```solidity
// Example: Simple wallet
contract MyWallet {
    event LogReceive(uint amount, string message);

    // Triggered when someone sends Ether directly
    receive() external payable {
        emit LogReceive(msg.value, "Ether received via receive()");
    }
}
```

👉 If Alice sends **1 ETH** directly to the contract address,
`receive()` will run and log:
`1 ETH received via receive()`

---

### ❓ `fallback()`

* Runs when:

  1. Someone calls a function that does **not exist**.
  2. Someone sends **data** that doesn’t match any function.
* Can also be `payable` to accept Ether.

```solidity
// Example: Handle unknown function calls
contract MyFallback {
    event LogFallback(uint amount, string message);

    fallback() external payable {
        emit LogFallback(msg.value, "Fallback triggered!");
    }
}
```

👉 If Bob tries to call `fooBar()` (which doesn’t exist),
or sends Ether **with data**,
then `fallback()` will run and log:
`Fallback triggered!`

---

### 🚨 Example Using Both Together

```solidity
contract EtherReceiver {
    event Received(string func, uint amount);

    // Case 1: plain Ether (no data)
    receive() external payable {
        emit Received("receive", msg.value);
    }

    // Case 2: unknown function OR data
    fallback() external payable {
        emit Received("fallback", msg.value);
    }
}
```

---

### 🎯 How it Works in Practice

1. **Send plain Ether** → `receive()` runs.
2. **Send Ether with random data** → `fallback()` runs.
3. **Call an unknown function** → `fallback()` runs.
4. **No `receive()` but `fallback()` exists** → `fallback()` also handles plain Ether.

---

👉 Think of it like this:

* `receive()` = **Dedicated mailbox for plain money transfers** 💸
* `fallback()` = **Catch-all box** 📦 (unknown calls + weird transfers)

### 🔄 **Function Overriding**

In Solidity, **derived contracts** can **override functions** from base contracts.

* The base function must be marked `virtual`.
* The overriding function must be marked `override`.

```solidity
// Base contract
contract Animal {
    function speak() public pure virtual returns (string memory) {
        return "Some sound";
    }
}

// Derived contract
contract Dog is Animal {
    function speak() public pure override returns (string memory) {
        return "Woof!";
    }
}

// Derived contract
contract Cat is Animal {
    function speak() public pure override returns (string memory) {
        return "Meow!";
    }
}
```

📌 **Key Points**

* `virtual` → allows a function to be overridden.
* `override` → indicates the derived contract is replacing the base function.
* Useful for **polymorphism** → different child contracts can provide **custom implementations** of the same function.

# 📘 If-Else Statements in Solidity

## 🔹 Introduction

In Solidity, `if-else` statements are used to make decisions in a smart contract. They allow developers to control the **flow of execution** by running certain code only if a condition is true. This is similar to conditional statements in other programming languages such as C, C++, or Java.

---

## 🔹 Syntax

```solidity
if (condition) {
    // code executes if condition is true
} else {
    // code executes if condition is false
}
```

For multiple conditions:

```solidity
if (condition1) {
    // executes if condition1 is true
} else if (condition2) {
    // executes if condition2 is true
} else {
    // executes if none of the above are true
}
```

---

## 🔹 Rules in Solidity

1. The condition inside `if` must evaluate to a **boolean** (`true` or `false`).
2. Solidity does **not** allow implicit conversions (e.g., you cannot just write `if(x)` like in Python). You must write something like `if(x > 0)`.
3. The `else` block is optional.
4. For input validation or reverting transactions, Solidity often uses `require`, `assert`, and `revert` instead of long `if-else` chains.

---

## 🔹 Examples

### ✅ Example 1: Voting Eligibility

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract Voting {
    function checkAge(uint _age) public pure returns (string memory) {
        if (_age >= 18) {
            return "Eligible to vote ✅";
        } else {
            return "Not eligible to vote ❌";
        }
    }
}
```

👉 If `_age = 20` → `"Eligible to vote ✅"`
👉 If `_age = 15` → `"Not eligible to vote ❌"`

---

### ✅ Example 2: Grading System

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract Grading {
    function getGrade(uint score) public pure returns (string memory) {
        if (score >= 90) {
            return "Grade: A";
        } else if (score >= 75) {
            return "Grade: B";
        } else if (score >= 50) {
            return "Grade: C";
        } else {
            return "Grade: F";
        }
    }
}
```

👉 `getGrade(92)` → `"Grade: A"`
👉 `getGrade(68)` → `"Grade: C"`

---

## 🔹 Best Practices

* Use `if-else` for logic that **changes behavior** depending on input.
* Use `require()` or `revert()` for **validations and error handling**, since they return unused gas and provide error messages.
* Keep conditions simple to reduce **gas consumption** and improve contract readability.

---

✅ **In short:**
`if-else` statements in Solidity allow contracts to make logical decisions. They are essential for handling different scenarios, but for validations and security, `require` and `revert` are generally preferred.

## ⚠️ **7. Error Handling in Solidity**

Solidity provides multiple ways to handle errors and ensure the safety and correctness of smart contract execution. These include:

* `require`
* `assert`
* `revert`

Understanding and using these correctly is **critical for contract security and gas efficiency**.

---


## ✅ A. `require` – Input Validation & Access Control

### 🧠 **Purpose**

The `require` statement is like a **checkpoint** in your code.
It makes sure that certain **conditions are true** before continuing execution.

* If the condition is **true** → the code continues normally.
* If the condition is **false** → the function **stops immediately**, any changes made so far are **undone (reverted)**, and the remaining gas is **refunded** to the caller.

Think of it like a **security guard at the gate**:
👉 If you meet the rules, you may enter.
👉 If you don’t, the system closes the gate and resets everything.

---

### 🔧 **Basic Syntax**

```solidity
require(condition, "Error message");
```

* `condition`: A statement that must be true.
* `"Error message"`: Optional, explains why the transaction failed.

---

### 📌 **Example**

```solidity
function transfer(uint amount) public {
    require(amount > 0, "Amount must be greater than 0");
    // If this line runs, it means 'amount > 0'
    // Continue with transfer logic...
}
```

👉 If `amount = 0`, the transaction is reverted and the message
`"Amount must be greater than 0"` is shown.

---

### 📌 **Common Use Cases**

1. **Ensuring Only the Right Caller Executes a Function**

   ```solidity
   require(msg.sender == owner, "Only the owner can call this function");
   ```

   ✅ Only the contract owner can access this.

---

2. **Validating Inputs**

   ```solidity
   require(_price > 0, "Price must be positive");
   ```

   ✅ Prevents invalid values from entering the system.

---

3. **Checking User Balances Before Withdrawal or Transfer**

   ```solidity
   require(balance[msg.sender] >= amount, "Insufficient balance");
   ```

   ✅ Ensures users cannot withdraw more than they have.

---

4. **Protecting Against Invalid States**

   ```solidity
   require(!paused, "Contract is paused");
   ```

   ✅ Stops actions if the contract is temporarily disabled.

---

### 🚨 What Happens on Failure?

* **Transaction Reverts**: All changes are undone.
* **Error Message Shown**: Helps users understand the reason.
* **Gas Refund**: Remaining unused gas is sent back.

---

### 🎯 **Key Points to Remember**

* Use `require` for **inputs and access control checks**.
* Place it at the **beginning of functions** to block invalid calls early.
* Always provide a **clear error message** → makes debugging and user experience better.


---

## ✅ B. `assert` – Internal Consistency Check

### 🧠 **Purpose**

The `assert` statement is like a **final lock** inside your contract.
It is **not for user input checks** (that’s `require`).
Instead, it’s used to confirm that your **internal program logic** is working correctly.

* If the condition is **true** → execution continues.
* If the condition is **false** → the transaction **reverts**, but unlike `require`, **all the gas is consumed** (⚠️ no refund).

👉 Think of it as a **self-check alarm** for the smart contract: if it rings, it means something is seriously wrong.

---

### 🔧 **Basic Syntax**

```solidity
assert(condition);
```

* No error message is provided (in most Solidity versions).
* It’s used for checks that should **never fail under normal circumstances**.

---

### 📌 **Example**

```solidity
function safeMath(uint a, uint b) public pure returns (uint) {
    uint result = a + b;
    assert(result >= a); // Check for overflow
    return result;
}
```

Here:

* If `result < a`, something is **logically impossible** (overflow bug).
* `assert` catches this unexpected situation.

---

### 📌 **Common Use Cases**

1. **Internal State Invariants**
   Rules about your contract’s state that must always hold true.

   ```solidity
   assert(totalSupply == balances[owner] + balances[user]);
   ```

---

2. **Overflow / Underflow Protection**
   Before Solidity 0.8 (which added automatic checks), developers used `assert` to prevent math errors.

   ```solidity
   uint c = a + b;
   assert(c >= a);
   ```

---

3. **Post-conditions**
   Ensuring the final state after a function matches your expectations.

   ```solidity
   function withdraw(uint amount) public {
       uint oldBalance = balance[msg.sender];
       balance[msg.sender] -= amount;
       assert(balance[msg.sender] == oldBalance - amount);
   }
   ```

---

### 🚨 What Happens on Failure?

* **Transaction is reverted**.
* **All gas is lost** (⚠️ no refund, unlike `require`).
* Indicates a **critical error** in the code, not user misuse.

---

### 🎯 **Key Points to Remember**

* ✅ Use `require` for **inputs & user checks**.
* ✅ Use `assert` only for **internal errors** and **logic that must never break**.
* ⚠️ Failing `assert` means your contract has a **bug**.

---

## ✅ C. `revert` – Custom Error Handling

### 🧠 **Purpose**

The `revert` statement **manually stops execution** and undoes any changes made so far.
It’s especially useful when you want **more control** over how and when to throw errors.

👉 Think of it as a **manual emergency brake** that you can pull whenever needed.

* **All state changes are reverted** (rolled back).
* **Unused gas is refunded** (like `require`).
* You can provide **custom error messages** or even **custom error types** (Solidity ≥ 0.8.4).

---

### 🔧 **Basic Syntax**

```solidity
revert("Error message here");
```

or with **custom errors**:

```solidity
error ErrorName();

if (condition) revert ErrorName();
```

---

### 📌 **Example 1 – Simple Revert**

```solidity
function withdraw(uint amount) public {
    if (amount > balance[msg.sender]) {
        revert("Not enough balance!");
    }
    balance[msg.sender] -= amount;
}
```

Here:

* If a user tries to withdraw more than they have → `revert` undoes the function and shows `"Not enough balance!"`.

---

### 📌 **Example 2 – Custom Errors (Gas-Efficient)**

Since Solidity 0.8.4, you can define **custom error types**, which save gas compared to long string messages.

```solidity
error Unauthorized();

function restrictedAction() public {
    if (msg.sender != owner) revert Unauthorized();
}
```

👉 Instead of wasting storage on long error strings, the contract emits a short, efficient identifier.

---

### 📌 **When to Use `revert`**

1. **Inside `if-else` Blocks**
   When logic needs multiple conditions.

   ```solidity
   if (x < 10) revert("x too small");
   else if (x > 100) revert("x too large");
   ```

---

2. **Custom Error Types**
   For **gas-saving** and **cleaner code**.

---

3. **Complex Validation Flows**
   When `require` becomes messy with multiple checks.

---

### 🚨 What Happens on Failure?

* **Transaction reverts** (all changes undone).
* **Unused gas refunded** (like `require`).
* Returns the **custom error** (string or defined error type).

---

### 🎯 **Key Points to Remember**

* ✅ Use `require` → simple conditions & input checks.
* ✅ Use `assert` → internal logic checks that must **never fail**.
* ✅ Use `revert` → for **custom error handling**, especially in **complex logic** or when saving gas with **custom errors**.


---

### 🆚 Summary Table: `require` vs `assert` vs `revert`

| Feature        | `require`                                                | `assert`                 | `revert`                     |
| -------------- | -------------------------------------------------------- | ------------------------ | ---------------------------- |
| Use Case       | Input validation, access check                           | Internal bugs/invariants | Custom error logic / nesting |
| Refund Gas     | ✅ Yes                                                    | ❌ No                     | ✅ Yes                        |
| Custom Msg     | ✅ Yes                                                    | ✅ (rarely used)          | ✅ Yes                        |
| Solidity ≥ 0.8 | Auto overflow/underflow checks make `assert` less needed |                          |                              |

---

### 🧪 Example Contract with All 3

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract ErrorExample {
    address public owner;
    mapping(address => uint) public balance;

    constructor() {
        owner = msg.sender;
    }

    function deposit() public payable {
        balance[msg.sender] += msg.value;
    }

    function withdraw(uint amount) public {
        require(amount > 0, "Amount must be > 0");
        require(balance[msg.sender] >= amount, "Insufficient funds");

        balance[msg.sender] -= amount;
        payable(msg.sender).transfer(amount);

        assert(balance[msg.sender] >= 0); // Should never be negative
    }

    function restricted() public {
        if (msg.sender != owner) {
            revert("Only owner can call this");
        }
        // do something only owner can do
    }
}
```

---

### ✅ What Students Should Practice

* Add `require` for all user-facing input checks.
* Use `revert` inside complex logic or `if-else` trees.
* Use `assert` only for *internal consistency* or math checks.

# 🔐 Get and Set Functions in Solidity

In Solidity, **Get and Set functions** are used to access and update the values of variables safely.  
They help you interact with the data stored in the smart contract.

---

## 📘 Definitions

| Term    | Meaning                                                 |
|---------|---------------------------------------------------------|
| `set`   | Function that **writes** or **updates** a variable.     |
| `get`   | Function that **reads** a variable's value.             |

---

## 🧠 Real-Life Analogy

Imagine a **locker** in a bank:

- 📝 `setPIN(1234)` → You **set the password** of the locker.
- 🔍 `getPIN()` → You **check** what the current password is.

---

## ✅ Why Use Get/Set?

- Helps organize your code
- Adds security (you can allow only certain users to update values)
- Makes the contract easier to interact with

---

## 🧪 What You Will Learn Below

1. `uint`: Numbers like age, score
2. `string`: Text like names
3. `bool`: True/False status
4. `address`: Wallet addresses
5. `struct`: Group of related data
6. `mapping`: Key-value store for data like balances

---

## ✅ Full Solidity Example

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

/// @title Get and Set Examples for Beginners
/// @notice Demonstrates how to use getters and setters for different data types

contract GetSetExamples {

    // 1️⃣ UINT: Example - Age
    uint private age;

    // Set age
    function setAge(uint _age) public {
        age = _age;
    }

    // Get age
    function getAge() public view returns (uint) {
        return age;
    }

    // 2️⃣ STRING: Example - Name
    string private name;

    function setName(string memory _name) public {
        name = _name;
    }

    function getName() public view returns (string memory) {
        return name;
    }

    // 3️⃣ BOOL: Example - Active status
    bool private isActive;

    function setIsActive(bool _status) public {
        isActive = _status;
    }

    function getIsActive() public view returns (bool) {
        return isActive;
    }

    // 4️⃣ ADDRESS: Example - Wallet address
    address private user;

    function setUser(address _user) public {
        user = _user;
    }

    function getUser() public view returns (address) {
        return user;
    }

    // 5️⃣ STRUCT: Group related data
    struct Profile {
        string username;
        uint balance;
    }

    Profile private myProfile;

    function setProfile(string memory _username, uint _balance) public {
        myProfile = Profile(_username, _balance);
    }

    function getProfile() public view returns (string memory, uint) {
        return (myProfile.username, myProfile.balance);
    }

    // 6️⃣ MAPPING: Key-value pairs (like a dictionary)
    mapping(address => uint) private balances;

    function setMyBalance(uint _amount) public {
        balances[msg.sender] = _amount;
    }

    function getMyBalance() public view returns (uint) {
        return balances[msg.sender];
    }
}
```

## 1️⃣ **Gas**

* Gas = the **unit of measurement** for computational work in Ethereum.
* Every operation in Solidity (adding numbers, storing data, sending Ether) costs some **gas units**.
* Example:

  * Adding two numbers → \~3 gas.
  * Writing to storage → \~20,000 gas.

Think of **gas like fuel** ⛽: the more complex your contract execution, the more gas you need.

---

## 2️⃣ **Execution Cost**

* This is the **actual amount of gas used** by your transaction.
* Formula:

  $$
  \text{Execution Cost} = \text{Gas Used} \times \text{Gas Price}
  $$
* Example:

  * If your function uses **30,000 gas** and gas price = **20 gwei**,

    $$
    30,000 \times 20 \text{ gwei} = 600,000 \text{ gwei} = 0.0006 ETH
    $$
* Execution cost = **the fuel you actually burned**.

---

## 3️⃣ **Transaction Cost**

* Transaction cost = **Execution Cost + Base Transaction Fee**.
* Every Ethereum transaction has a **base fee** (\~21,000 gas just for sending a transaction).
* Formula:

  $$
  \text{Transaction Cost} = (21,000 + \text{Gas Used}) \times \text{Gas Price}
  $$
* Example:

  * You call a function using **30,000 gas**.
  * Transaction cost = (21,000 + 30,000) × 20 gwei = **1,020,000 gwei = 0.00102 ETH**.

So **transaction cost** is always slightly higher than execution cost.

---

## 📊 **Quick Comparison Table**

| Term                 | Meaning                            | Example Analogy                          |
| -------------------- | ---------------------------------- | ---------------------------------------- |
| **Gas**              | Unit of work (like liters of fuel) | "This journey needs 30 liters."          |
| **Execution Cost**   | Actual gas used × gas price        | "You drove 30 liters × price per liter." |
| **Transaction Cost** | Execution cost + base fee          | "You paid fuel + toll tax."              |

---

✅ **Summary**

* **Gas** = abstract unit (measures work).
* **Execution Cost** = how much ETH you burned for actual computation.
* **Transaction Cost** = execution cost + base 21,000 gas fee.

---


Terminology gets tricky. Let me clarify some points:

---

## 1️⃣ **Gas Itself**

* **Gas is not money.**
  Gas is just the **unit of measurement** for work done in Ethereum (like liters of fuel or CPU cycles).
* Example: *This function uses 30,000 gas units.*

So **gas itself is not the miner’s fee** — it’s just the quantity.

---

## 2️⃣ **Gas Price**

* Gas Price = how much you are willing to **pay per unit of gas** (in gwei or ETH).
* Example: *20 gwei per gas.*

---

## 3️⃣ **Fee to Miners / Validators**

* The **fee miners (PoW) or validators (PoS)** actually earn =

  $$
  \text{Gas Used} \times \text{Gas Price}
  $$
* This is the **execution cost**, and yes, it **goes to miners/validators** as their reward for processing your transaction.

---

## 4️⃣ **Transaction Fee Breakdown**

* **Base Fee** (introduced in EIP-1559) is **burned** (removed from circulation).
* **Priority Fee (tip)** goes to the miner/validator.
* Formula under EIP-1559 (current Ethereum):

  $$
  \text{Total Transaction Fee} = \text{Gas Used} \times (\text{Base Fee} + \text{Priority Fee})
  $$

So **not all of your gas payment goes to miners** anymore — part of it is burned.

---

## 📊 **Analogy**

* **Gas units** = number of liters of fuel your car uses.
* **Gas price** = cost per liter.
* **Execution cost** = liters × cost = what you actually pay.
* **Miner’s reward** = only part of this cost (priority fee), the rest (base fee) is destroyed.

## 📘 Week 6 - Class 4: Modifiers, Payments & Mini Project

### 🔐 What Are Modifiers in Solidity?

In Solidity, **modifiers** are like **guards** or **rules** you can attach to functions.
They:

* Control **who can call** the function
* Check **conditions before running** the function
* Help avoid repeating the same `require` checks in every function

👉 Think of them as **reusable security filters** for your functions.

---

### ✅ Basic Syntax

```solidity
modifier modifierName {
    // condition(s)
    _;
}
```

* `modifierName` → name you give to the rule
* `_` (underscore) → this is where the function’s code will run **if conditions are true**

---

### 🔒 Example 1: `onlyOwner` (Access Control)

Most contracts need some functions only the **owner (admin)** can run.

```solidity
pragma solidity ^0.8.0;

contract MyContract {
    address public owner;

    constructor() {
        owner = msg.sender; // whoever deploys is the owner
    }

    // modifier for owner-only functions
    modifier onlyOwner() {
        require(msg.sender == owner, "Not the contract owner!");
        _;
    }

    // function restricted to owner
    function changeOwner(address newOwner) public onlyOwner {
        owner = newOwner;
    }
}
```

🧠 **Explanation**:

* `onlyOwner` checks if `msg.sender` (caller) is the owner.
* If not → transaction **fails**.
* If yes → function runs normally.

---

### 🧮 Example 2: Input Validation

You can make modifiers to **check inputs** instead of writing `require` everywhere.

#### 🔧 `validAmount` Modifier

```solidity
modifier validAmount(uint amount) {
    require(amount > 0, "Amount must be greater than 0");
    _;
}

function deposit(uint amount) public validAmount(amount) {
    // deposit logic
}
```

#### 🔧 `notPaused` Modifier

```solidity
bool public paused = false;

modifier notPaused() {
    require(!paused, "Contract is paused");
    _;
}

function buyTokens() public notPaused {
    // token buying logic
}
```

---

### 🛠 Common Modifier Patterns

| Modifier Name | Purpose                            |
| ------------- | ---------------------------------- |
| `onlyOwner`   | Restrict to admin/owner            |
| `notPaused`   | Pause contract for safety          |
| `validAmount` | Check amounts are positive         |
| `onlyBuyer`   | Restrict to specific role          |
| `nonZero`     | Prevent `0` or empty address input |

---

### ⚙️ Using Multiple Modifiers

You can stack modifiers together for stronger checks:

```solidity
function buyItem(uint amount)
    public
    onlyOwner
    validAmount(amount)
{
    // function logic here
}
```

👉 First `onlyOwner` runs → if it passes, then `validAmount` runs → if both pass, the function executes.

---

### 🔐 Why Modifiers Are Important?

- ✔ Make code **clean & reusable**

## 💰 **Payable Functions & Ether Transfers**

In Solidity, we often deal with **real money (Ether)**.
To safely move Ether in and out of contracts, Solidity gives us **payable functions** and a few transfer methods.

---

### ✅ What is `payable`?

* `payable` means: **this function or address can accept Ether**.
* If you try sending Ether to a non-payable function → the transaction will **fail**.

👉 Think of it as adding a **“money slot”** to your function.

---

### 📥 Receiving Ether

```solidity
pragma solidity ^0.8.0;

contract Wallet {
    address public owner;

    constructor() {
        owner = msg.sender; // whoever deploys becomes owner
    }

    // ✅ Anyone can deposit Ether
    function deposit() public payable {}

    // ✅ Check how much Ether is inside the contract
    function getBalance() public view returns (uint) {
        return address(this).balance;
    }
}
```

🧠 **Key Points**

* `msg.value` → Ether sent with a function call.
* `address(this).balance` → total Ether in the contract.
* In Remix/MetaMask: set **Value** field > click **deposit()**.

---

### 💸 Sending Ether Out

There are 3 ways to send Ether:

| Method     | Auto Revert if Fail | Gas Limit        | Safety                            |
| ---------- | ------------------- | ---------------- | --------------------------------- |
| `transfer` | ✅ Yes               | 2300 gas         | ✅ Safe, but limited               |
| `send`     | ❌ No                | 2300 gas         | ❌ Unsafe (must check result)      |
| `call`     | ❌ No (must check)   | Forwards all gas | ✅ Best practice (flexible & safe) |

---

#### Example 1: Using `transfer()`

```solidity
function withdraw() public {
    require(msg.sender == owner, "Not owner");
    payable(msg.sender).transfer(address(this).balance);
}
```

* Sends entire contract balance to owner.
* Automatically reverts if it fails.

---

#### Example 2: Using `call()` (Preferred)

```solidity
function safeWithdraw() public {
    require(msg.sender == owner, "Not owner");

    (bool success, ) = payable(msg.sender).call{value: address(this).balance}("");
    require(success, "Transfer failed");
}
```

* `call` is now recommended over `transfer`.
* Must **check success** manually.

---

### 🔐 Restricting Withdrawals

We usually only want the **owner** to withdraw.

```solidity
modifier onlyOwner() {
    require(msg.sender == owner, "Not authorized");
    _;
}

function withdrawOwner() public onlyOwner {
    (bool success, ) = payable(msg.sender).call{value: address(this).balance}("");
    require(success, "Withdraw failed");
}
```

   * `deposit()` (payable)
   * `getBalance()` (view)
   * `withdraw()` using `call()`
2. Deploy in Remix → select **Injected Provider (MetaMask)** or **JavaScript VM**.
3. Try:

   * Sending Ether with `deposit()` → works ✅
   * Calling a non-payable function with Ether → fails ❌

### ⚠️ Best Practices

- ✔ Always check return values (with `call`)
- ✔ Avoid using `send`
- ✔ Protect withdrawals with `onlyOwner`
- ✔ Prefer `call` over `transfer`
- ✔ Set owner in constructor instead of hardcoding

## 🧮 What are `decimals`?

* ERC-20 tokens usually don’t store balances in *whole numbers*.
* Instead, they store balances in the **smallest unit** (like cents for a dollar).

Example analogy:

* **Dollars → 2 decimal places (cents)**
* **Ethereum & most ERC-20 tokens → 18 decimal places**

So when your contract has:

```solidity
uint8 public decimals = 18;
```

It means:
👉 `1 token` in human view = `1 * 10^18` in the contract’s internal storage.

---

## 📊 Example with MyToken

* You deploy with `_initialSupply = 1000`.
* Inside the constructor:

  ```solidity
  totalSupply = _initialSupply * (10 ** uint(decimals));
  ```
* With `decimals = 18`:

  ```
  totalSupply = 1000 * 10^18
              = 1000000000000000000000
  ```

So Remix will show `1000000000000000000000` instead of `1000`.

But logically it still means **1000 tokens with 18 decimal precision**.

---

## 💡 Why 18?

* Ethereum itself uses **18 decimals** (1 ETH = 1e18 wei).
* Most ERC-20 tokens follow this standard for compatibility.

---

## ✅ How to Handle in Remix

* If you want to **send 5 tokens**, you must enter:

  ```
  5 * 10^18 = 5000000000000000000
  ```
* If you don’t want this headache while testing → change:

  ```solidity
  uint8 public decimals = 0;
  ```

  Then:

  * Total supply = exactly what you enter (e.g., 1000).
  * To transfer 5 tokens → just write `5`.

---

⚡ So:

* `decimals = 18` → realistic (like ETH, USDT).
* `decimals = 0` → easier for Remix classroom demos.

# Creating Your Own Cryptocurrency (ERC-20 Token)

---

### 🔑 What is ERC-20?

* **ERC-20** = Ethereum Request for Comments #20.
* It’s the **most common token standard** on Ethereum.
* It defines a **set of rules** so that:

  * Wallets (like MetaMask) can recognize tokens.
  * Exchanges (like Binance, Uniswap) can list and trade them.
  * Other smart contracts (like DeFi apps) can interact with them.

👉 Think of ERC-20 as a **blueprint/template for making your own coins**.
Examples: **USDT (Tether), DAI, UNI, LINK** — all are ERC-20 tokens.

---

## 🧱 Core Features of ERC-20

### 1️⃣ `totalSupply()`

**Purpose:** Shows the total number of tokens created.

📌 Example:

* If you set `totalSupply = 1,000,000` → that’s the maximum number of tokens that exist.

---

### 2️⃣ `balanceOf(address)`

**Purpose:** Shows how many tokens an account has.

📌 Example:

* If you call `balanceOf(0x123...)` and get `500`, that wallet owns 500 tokens.

---

### 3️⃣ `transfer(to, amount)`

**Purpose:** Lets you **send tokens** directly to another account.

📌 Example:

* `transfer(Alice, 100)` → moves 100 tokens from you to Alice.
* Used for simple payments, rewards, or sending tokens to friends.

---

### 4️⃣ `approve(spender, amount)`

**Purpose:** Allows you to **give permission** to another account (spender) to spend your tokens.

📌 Example:

* You approve a **decentralized exchange (DEX)** to spend `500` of your tokens.
* The DEX can then move those tokens from your wallet during a trade.

---

### 5️⃣ `transferFrom(from, to, amount)`

**Purpose:** Lets someone (the spender) **transfer tokens on behalf of another account**, after approval.

📌 Example:

* You approve Uniswap to spend `500` tokens.
* When you trade, Uniswap calls `transferFrom(you → liquidity pool, 500)`.

---

### 6️⃣ `allowance(owner, spender)`

**Purpose:** Shows **how many tokens** a spender is still allowed to use from the owner’s account.

📌 Example:

* If you approved Bob for `200` tokens, and he already spent `50`, then:

  * `allowance(you, Bob)` = `150`.

---

## ⚡ Extra Features We Added

To make the project more powerful, we added some **extra functions** beyond the basic ERC-20.

---

### 🔥 `mint(address, amount)`

* Owner can **create new tokens** and give them to an address.
* Increases the total supply.
* Useful for **rewards, fundraising, or token inflation**.

📌 Example: Owner mints 100 tokens to Alice → Alice’s balance increases, totalSupply goes up.

---

### 🗑 `burn(amount)`

* Any holder can **destroy their tokens** permanently.
* Decreases the total supply.
* Useful for **deflationary tokens** (like Shiba Inu burns supply to increase scarcity).

📌 Example: You burn 50 tokens → your balance goes down, totalSupply decreases.

---

### ⏸ `pause()` / ▶️ `unpause()`

* Only the owner can pause the contract.
* When paused → transfers, mint, burn, etc. are **blocked**.
* Useful in emergencies (e.g., hack, bug, or upgrade).

📌 Example:

* Admin pauses the contract → no one can transfer tokens until `unpause()`.

---

### 👑 `onlyOwner` Modifier

* Restricts some functions (like `mint`, `pause`) to **the contract owner only**.
* Prevents random people from abusing critical functions.

📌 Example:

* Only the owner can call `mint()` or `pause()`.
* Students will see how modifiers improve **security** and **code reusability**.

---

## 🎯 Why These Features Matter

* **Basic ERC-20 (6 core functions)** → ensures compatibility with wallets, exchanges, and dApps.
* **Mint** → lets projects raise tokens over time.
* **Burn** → lets tokens become scarce (increasing value).
* **Pause/Unpause** → adds security, like an emergency stop button.
* **onlyOwner** → enforces admin control.

## 📝 Full ERC-20 Token

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

/**
 * @title MyToken (ERC-20 Example)
 * @dev A simple ERC-20 token with mint, burn, pause/unpause features.
 */
contract MyToken {
    // -------------------------------
    // 🔹 Token Basic Info
    // -------------------------------
    string public name = "MyToken";       // Token name
    string public symbol = "MTK";         // Token symbol
    uint8 public decimals = 18;           // Standard: 18 decimals
    uint public totalSupply;              // Total tokens in circulation

    // -------------------------------
    // 🔹 Mappings
    // -------------------------------
    mapping(address => uint) public balanceOf;                      // Track token balances
    mapping(address => mapping(address => uint)) public allowance;  // Track approvals (who can spend on behalf of whom)

    // -------------------------------
    // 🔹 Owner & Pause state
    // -------------------------------
    address public owner;   // Contract deployer is the owner (admin)
    bool public paused;     // If true → contract is paused (no transfers)

    // -------------------------------
    // 🔹 Events (important for dApps & block explorers)
    // -------------------------------
    event Transfer(address indexed from, address indexed to, uint value);
    event Approval(address indexed owner, address indexed spender, uint value);
    event Paused(address account);
    event Unpaused(address account);
    event Mint(address indexed to, uint value);
    event Burn(address indexed from, uint value);

    // -------------------------------
    // 🔹 Constructor
    // -------------------------------
    constructor(uint _initialSupply) {
        owner = msg.sender;  // Deployer becomes owner
        totalSupply = _initialSupply * (10 ** uint(decimals));  // Adjust for decimals
        balanceOf[msg.sender] = totalSupply; // Assign all tokens to owner initially
        emit Transfer(address(0), msg.sender, totalSupply); // Emit event (from "zero address" = mint)
    }

    // -------------------------------
    // 🔹 Modifiers
    // -------------------------------
    modifier onlyOwner() {
        require(msg.sender == owner, "Not the contract owner!");
        _;
    }

    modifier notPaused() {
        require(!paused, "Contract is paused!");
        _;
    }

    // -------------------------------
    // 🔹 ERC-20 Core Functions
    // -------------------------------

    // Transfer tokens to another address
    function transfer(address _to, uint _value) public notPaused returns (bool success) {
        require(balanceOf[msg.sender] >= _value, "Not enough tokens");
        balanceOf[msg.sender] -= _value;   // Deduct from sender
        balanceOf[_to] += _value;          // Add to receiver
        emit Transfer(msg.sender, _to, _value); // Log event
        return true;
    }

    // Approve another address to spend tokens on your behalf
    function approve(address _spender, uint _value) public notPaused returns (bool success) {
        allowance[msg.sender][_spender] = _value;  // Set approval
        emit Approval(msg.sender, _spender, _value);
        return true;
    }

    // Transfer tokens on behalf of someone (requires approval first)
    function transferFrom(address _from, address _to, uint _value) public notPaused returns (bool success) {
        require(balanceOf[_from] >= _value, "Not enough tokens");
        require(allowance[_from][msg.sender] >= _value, "Not approved to spend this amount");

        balanceOf[_from] -= _value;             // Deduct from sender
        balanceOf[_to] += _value;               // Add to recipient
        allowance[_from][msg.sender] -= _value; // Reduce spender’s allowance

        emit Transfer(_from, _to, _value);
        return true;
    }

    // -------------------------------
    // 🔹 Extra Features
    // -------------------------------

    // Mint new tokens (only owner)
    function mint(address _to, uint _value) public onlyOwner notPaused {
        uint amount = _value * (10 ** uint(decimals));
        totalSupply += amount;               // Increase supply
        balanceOf[_to] += amount;            // Give to recipient
        emit Mint(_to, amount);
        emit Transfer(address(0), _to, amount); // ERC-20 standard (from 0 = mint)
    }

    // Burn tokens (destroy them forever)
    function burn(uint _value) public notPaused {
        uint amount = _value * (10 ** uint(decimals));
        require(balanceOf[msg.sender] >= amount, "Not enough tokens to burn");
        balanceOf[msg.sender] -= amount;  // Remove from sender
        totalSupply -= amount;            // Decrease supply
        emit Burn(msg.sender, amount);
        emit Transfer(msg.sender, address(0), amount); // To 0 = burn
    }

    // Pause the contract (onlyOwner)
    function pause() public onlyOwner {
        paused = true;
        emit Paused(msg.sender);
    }

    // Unpause the contract (onlyOwner)
    function unpause() public onlyOwner {
        paused = false;
        emit Unpaused(msg.sender);
    }
}
```

## 🔹 Basic Token Info

👉 Example: If `totalSupply = 1000 * 10^18`, it means **1000 tokens** with 18 decimal precision.

---

## 🔹 Mappings (Token Balances & Approvals)

* `balanceOf[address]` → Keeps track of **how many tokens each address owns**.
* `allowance[owner][spender]` → Shows how many tokens the **owner approved** the **spender** to use on their behalf.

👉 Example: If Alice approves Bob for `50`, then Bob can spend up to `50` tokens from Alice’s balance.

---

## 🔹 Owner & Pause State

* `owner` → The account that deployed the contract (admin).
* `paused` → Boolean switch. If `true`, most functions stop working (emergency stop).

👉 Useful if there’s a hack or bug → you can pause the token until it’s fixed.

---

## 🔹 Events

Events are like **logs** that wallets, block explorers (like Etherscan), and dApps use to track actions.

* `Transfer` → Fired when tokens move between accounts.
* `Approval` → Fired when someone approves another address to spend tokens.
* `Paused` / `Unpaused` → Fired when the contract is paused/unpaused.
* `Mint` / `Burn` → Fired when tokens are created or destroyed.

---

## 🔹 Constructor

Runs **once at deployment**.

* Sets the deployer as the `owner`.
* Creates the initial supply and assigns it to the deployer.
* Emits a `Transfer` from the "zero address" (standard ERC-20 way of recording minting).

👉 Example: If you deploy with `1000`, you (the owner) start with 1000 tokens.

---

## 🔹 Modifiers

Modifiers are **reusable checks** that run before a function.

* `onlyOwner` → Restricts functions to the contract owner (admin).
* `notPaused` → Ensures the contract is not paused.

👉 They save code repetition and improve security.

---

## 🔹 ERC-20 Core Functions

### 1️⃣ `transfer(address _to, uint _value)`

* Send tokens from **your account** to another account.
* Requires you to have enough tokens.
* Updates balances and emits a `Transfer` event.

👉 Example: Alice calls `transfer(Bob, 100)`. Alice’s balance decreases by 100, Bob’s increases by 100.

---

### 2️⃣ `approve(address _spender, uint _value)`

* Lets you give **permission** to another address to spend your tokens.
* Sets the allowance value.
* Emits an `Approval` event.

👉 Example: Alice approves Uniswap (a DEX) to spend 500 tokens on her behalf.

---

### 3️⃣ `transferFrom(address _from, address _to, uint _value)`

* Used by a **spender** to move tokens **on behalf of someone else**, after approval.
* Checks both the balance and the allowance.
* Deducts from allowance as it’s used.
* Emits a `Transfer` event.

👉 Example: Bob has approval for 50 tokens from Alice. Bob calls `transferFrom(Alice, Bob, 50)` → tokens move from Alice to Bob.

---

## 🔹 Extra Features

### 🔥 `mint(address _to, uint _value)`

* **Minting = Creating new tokens**.
* Only the owner can mint.
* Increases `totalSupply` and credits the tokens to the `_to` address.
* Emits `Mint` + `Transfer` (from zero address).

👉 Example: Owner mints 200 tokens to Alice → totalSupply goes up, Alice’s balance increases.

---

### 🗑 `burn(uint _value)`

* **Burning = Destroying tokens permanently**.
* Decreases both the sender’s balance and `totalSupply`.
* Emits `Burn` + `Transfer` (to zero address).

👉 Example: Alice burns 50 tokens → her balance goes down, supply decreases → tokens are gone forever.

---

### ⏸ `pause()` / ▶️ `unpause()`

* Only owner can call these.
* `pause()` → sets `paused = true`. Prevents transfers, minting, and burning.
* `unpause()` → sets `paused = false`. Contract resumes normal operation.

👉 Example: If a bug is found, owner pauses the token to stop all activity. Once fixed, owner unpauses.

# 🎨 Non-Fungible Tokens (NFTs)


---



## 🔑 Definition

An **NFT (Non-Fungible Token)** is a type of digital asset stored on a blockchain that represents **ownership of something unique**.

* **Non-fungible** means it **cannot be replaced or exchanged on a one-to-one basis** with another asset.
* Each NFT has a unique **token ID** and usually links to metadata (information like image, description, attributes).

---

## 📦 Difference Between Fungible and Non-Fungible

* **Fungible (ERC-20 tokens, like ETH, USDT):**

  * Interchangeable.
  * 1 ETH = 1 ETH (all tokens are the same).

* **Non-Fungible (ERC-721 tokens, like NFTs):**

  * Unique.
  * NFT #1 ≠ NFT #2 (each has its own identity, metadata, and traits).

---

## 🌍 Common Use Cases of NFTs

1. **Digital Art** – Ownership of unique artworks.
2. **Gaming** – In-game assets like swords, skins, or characters.
3. **Collectibles** – Digital trading cards, limited editions.
4. **Music & Media** – Songs, albums, videos as NFTs.
5. **Certificates & Tickets** – Event passes, diplomas, licenses.

---

## 📑 Metadata in NFTs

An NFT itself is just an ID on the blockchain. The **details** (name, description, image, traits) are stored in a **metadata JSON file**, usually on **IPFS**.

Example Metadata:

```json
{
  "name": "Golden Sword",
  "description": "A rare sword NFT from the fantasy game.",
  "image": "ipfs://Qm...hashOfImage",
  "attributes": [
    { "trait_type": "Attack Power", "value": "120" },
    { "trait_type": "Rarity", "value": "Legendary" }
  ]
}
```

👉 This metadata makes each NFT **unique** and gives it real-world meaning.

---




# 🔗 How NFTs and Smart Contracts Connect

* NFTs are created using **ERC-721 smart contracts**.
* The smart contract defines:

  1. **How NFTs are minted** (created).
  2. **How ownership is tracked** (each NFT has an owner).
  3. **How NFTs are transferred** (buy, sell, gift).
  4. **How metadata is linked** (via tokenURI → points to IPFS JSON).

---

# 🛠 NFT Lifecycle (Step by Step)

1. **Deploy Contract** → Smart contract is published on the blockchain.
2. **Mint NFT** → Contract creates NFT #0 and assigns it to an address.
3. **Attach Metadata** → NFT points to a JSON file (with name, image, description).
4. **Transfer/Sell** → NFT can be sent to another wallet.
5. **View in Wallet/Marketplace** → MetaMask, OpenSea, etc., read metadata and show the image/art.

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.20; // ✅ Compatible with OpenZeppelin v5.x

// -------------------------------
// 📦 Importing OpenZeppelin Libraries
// -------------------------------
import "@openzeppelin/contracts/token/ERC721/extensions/ERC721URIStorage.sol"; // ERC721 with storage-based metadata
import "@openzeppelin/contracts/access/Ownable.sol";                          // Ownership access control

/**
 * @title MyNFT
 * @dev ERC-721 Non-Fungible Token with metadata support (IPFS/HTTPS URIs).
 */
contract MyNFT is ERC721URIStorage, Ownable {
    uint public tokenCounter; // Counter to assign unique token IDs

    // -------------------------------
    // 🏗 Constructor
    // -------------------------------
    constructor() ERC721("MyNFTCollection", "MNFT") Ownable(msg.sender) {
        // "MyNFTCollection" = Collection Name
        // "MNFT"            = Token Symbol
        // msg.sender        = Deployer becomes contract owner
        tokenCounter = 0; // Start IDs from 0
    }

    // -------------------------------
    // 🖼 Minting Function
    // -------------------------------
    /**
     * @dev Mint a new NFT with metadata
     * @param recipient The wallet address that will receive the NFT
     * @param tokenURI  The URI pointing to metadata JSON (usually on IPFS or HTTPS)
     * @return newTokenId The ID of the newly minted NFT
     */
    function mintNFT(address recipient, string memory tokenURI)
        public
        onlyOwner // ✅ Only contract owner can mint
        returns (uint newTokenId)
    {
        newTokenId = tokenCounter;            // Current counter is the new Token ID
        _safeMint(recipient, newTokenId);     // Mint NFT safely to recipient
        _setTokenURI(newTokenId, tokenURI);   // Attach metadata URI to this NFT
        tokenCounter++;                       // Increment counter for next mint
    }

    // -------------------------------
    // 🔥 Burn Function
    // -------------------------------
    /**
     * @dev Burn (destroy) an existing NFT.
     * @param tokenId The ID of the token to burn.
     * Caller must own the token or be approved.
     */
    function burn(uint tokenId) public {
        // ✅ Use built-in check: only owner or approved address can burn
        require(_isApprovedOrOwner(_msgSender(), tokenId), "Not authorized to burn this token");
        _burn(tokenId); // ✅ Built-in OpenZeppelin burn
    }

    // -------------------------------
    // 🛠 Internal Helper
    // -------------------------------
    /**
     * @dev Expose the internal `_isApprovedOrOwner` from ERC721
     */
    function _isApprovedOrOwner(address spender, uint tokenId) internal view returns (bool) {
        address owner = ownerOf(tokenId);
        return (spender == owner || isApprovedForAll(owner, spender) || getApproved(tokenId) == spender);
    }
}

```

### 3. Contract Declaration

```solidity
contract MyNFT is ERC721URIStorage, Ownable {
    uint public tokenCounter;
```

* The contract is called `MyNFT`.

### 4. Constructor

```solidity
constructor() ERC721("MyNFTCollection", "MNFT") {
    tokenCounter = 0;
}
```

* Runs only once at deployment.
* Sets collection **name** = `"MyNFTCollection"` and **symbol** = `"MNFT"`.
* Starts `tokenCounter` at 0.

---

### 5. Mint Function

```solidity
function mintNFT(address recipient, string memory tokenURI)
    public
    onlyOwner
    returns (uint newTokenId)
```

* **Inputs**:

  * `recipient`: Address that will own the NFT.
  * `tokenURI`: IPFS/HTTP link to JSON metadata.
* **onlyOwner**: Only the contract owner can mint.
* **Returns**: The new token ID.

---

### 6. Mint Process

```solidity
newTokenId = tokenCounter;           // Use counter as token ID
_safeMint(recipient, newTokenId);    // Mint NFT and assign to recipient
_setTokenURI(newTokenId, tokenURI);  // Link NFT to metadata
tokenCounter += 1;                   // Prepare next ID
```

* Assigns NFT ID = current counter.
* Creates NFT safely (ensures recipient can handle NFTs).
* Attaches metadata JSON (which describes image, name, traits).
* Increments counter for the next mint.

# 🖼 Metadata Example

NFTs need metadata in **JSON format**, stored on IPFS (or HTTPS).

```json
{
  "name": "My First NFT",
  "description": "This is my first NFT created in Solidity!",
  "image": "ipfs://QmYourImageHash",
  "attributes": [
    { "trait_type": "Level", "value": "1" },
    { "trait_type": "Rarity", "value": "Legendary" }
  ]
}
```

* `name`: The NFT’s name.
* `description`: Details about it.
* `image`: Link to artwork (usually IPFS).
* `attributes`: Extra properties (used in games, collectibles).

# 🛠 How It Works in Practice

1. **Deploy contract** → You become the owner.
2. **Mint NFT** → Call `mintNFT(yourAddress, "ipfs://metadataCID")`.
3. **Token ID** → NFT #0 created, linked to metadata.
4. **Check ownership** → Call `ownerOf(0)` → returns your address.
5. **View NFT** → Add contract address + token ID in a wallet (e.g., MetaMask) or marketplace (e.g., OpenSea).
